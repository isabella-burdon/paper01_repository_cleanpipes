```python
# dependancies
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Arial'
```

```python
# data directories
SR_nopreproc_dir = '../profiled_data/SR_shotgun_noPreprocessing'
SR_filtered_dir = '../profiled_data/SR_shotgun'
LR_nopreprocess_dir = '../profiled_data/LR_shotgun_noPreprocessing'
LR_filtered_dir = '../profiled_data/LR_shotgun'
```

### Load data into list of dataframes

```python
# initiliase list of dfs
SR_nopreproc_df_list = []
SR_filtered_df_list = []
LR_nopreproc_df_list = []
LR_filtered_df_list = []

# SR_nopreproc_df_list
SR_nopreproc_files = os.listdir(SR_nopreproc_dir)
for file in SR_nopreproc_files:
    if file.endswith('.tsv'):
        df = pd.read_csv(SR_nopreproc_dir + '/' + file, sep='\t')
        SR_nopreproc_df_list.append(df)

# SR_filtered_df_list
SR_filtered_files = os.listdir(SR_filtered_dir)
for file in SR_filtered_files:
    if file.endswith('.tsv'):
        df = pd.read_csv(SR_filtered_dir + '/' + file, sep='\t')
        SR_filtered_df_list.append(df)

# LR_nopreproc_df_list
LR_nopreproc_files = os.listdir(LR_nopreprocess_dir)
for file in LR_nopreproc_files:
    if file.endswith('.tsv'):
        df = pd.read_csv(LR_nopreprocess_dir + '/' + file, sep='\t')
        LR_nopreproc_df_list.append(df)

# LR_filtered_df_list
LR_filtered_files = os.listdir(LR_filtered_dir)
for file in LR_filtered_files:
    if file.endswith('.tsv'):
        df = pd.read_csv(LR_filtered_dir + '/' + file, sep='\t')
        LR_filtered_df_list.append(df)
```

### Extract relevant data
Classified v unclassified reads

```python
# df with columns classified and unclassified, rows are the samples

nopreSR_classunclass_df = pd.DataFrame(columns=['Mock', 'Classified Reads', 'Unclassified Reads'])
filtSR_classunclass_df = pd.DataFrame(columns=['Mock', 'Classified Reads', 'Unclassified Reads'])
nopreLR_classunclass_df = pd.DataFrame(columns=['Mock', 'Classified Reads', 'Unclassified Reads'])
filtLR_classunclass_df = pd.DataFrame(columns=['Mock', 'Classified Reads', 'Unclassified Reads'])
SR_samples = ['A', 'B', 'C', 'E', 'F']
LR_samples = ['A', 'B', 'C', 'D', 'E', 'F']

# SR no preprocessing
for idx, value in enumerate(SR_samples):
    df = SR_nopreproc_df_list[idx]
    unclassified = df.loc[df['superkingdom'] == 'unclassified', 'fraction'].sum()
    classified = 1 - unclassified
    mock = value
    new_row = pd.DataFrame({'Mock': [mock],
                            'Classified Reads': [classified],
                            'Unclassified Reads': [unclassified]})
    nopreSR_classunclass_df = pd.concat([nopreSR_classunclass_df, new_row], 
                                   ignore_index=True)

# SR filtered
for idx, value in enumerate(SR_samples):
    df = SR_filtered_df_list[idx]
    unclassified = df.loc[df['superkingdom'] == 'unclassified', 'fraction'].sum()
    classified = 1 - unclassified
    mock = value
    new_row = pd.DataFrame({'Mock': [mock],
                            'Classified Reads': [classified],
                            'Unclassified Reads': [unclassified]})
    filtSR_classunclass_df = pd.concat([filtSR_classunclass_df, new_row], 
                                   ignore_index=True)

# LR no preprocessing
for idx, value in enumerate(LR_samples):
    df = LR_nopreproc_df_list[idx]
    unclassified = df.loc[df['superkingdom'] == 'unclassified', 'fraction'].sum()
    classified = 1 - unclassified
    mock = value
    new_row = pd.DataFrame({'Mock': [mock],
                            'Classified Reads': [classified],
                            'Unclassified Reads': [unclassified]})
    nopreLR_classunclass_df = pd.concat([nopreLR_classunclass_df, new_row], 
                                   ignore_index=True)
    
# LR filtered
for idx, value in enumerate(LR_samples):
    df = LR_filtered_df_list[idx]
    unclassified = df.loc[df['superkingdom'] == 'unclassified', 'fraction'].sum()
    classified = 1 - unclassified
    mock = value
    new_row = pd.DataFrame({'Mock': [mock],
                            'Classified Reads': [classified],
                            'Unclassified Reads': [unclassified]})
    filtLR_classunclass_df = pd.concat([filtLR_classunclass_df, new_row], 
                                   ignore_index=True)
```

### Visualise data
, bbox_to_anchor=(1.02, 1)

```python
# FIGURE 1 FUNCTION: BACTERIAL v CONTAMINANTS

def figure_1(nopreSR_classunclass_df, filtSR_classunclass_df, 
             nopreLR_classunclass_df, filtLR_classunclass_df, 
             background_colour, plot_colour, font_colour, line_colour,
             title, legend_labels, colour_map,
             plot_title1, x_labels1, x_title1, y_title1,
             plot_title2, x_labels2, x_title2, y_title2,
             plot_title3, x_labels3, x_title3, y_title3,
             plot_title4, x_labels4, x_title4, y_title4):
    
    # data
    nopreproc_SR_data = nopreSR_classunclass_df[legend_labels].values.astype(np.float64)
    filtered_SR_data = filtSR_classunclass_df[legend_labels].values.astype(np.float64)
    nopreproc_LR_data = nopreLR_classunclass_df[legend_labels].values.astype(np.float64)
    filtered_LR_data = filtLR_classunclass_df[legend_labels].values.astype(np.float64)
    data1 = nopreproc_SR_data * 100
    data2 = filtered_SR_data * 100
    data3 = nopreproc_LR_data * 100
    data4 = filtered_LR_data * 100

    datas = [data1, data2, data3, data4]
    plot_titles = [plot_title1, plot_title2, plot_title3, plot_title4]
    all_x_labels = [x_labels1, x_labels2, x_labels3, x_labels4]
    x_titles = [x_title1, x_title2, x_title3, x_title4]
    y_titles = [y_title1, y_title2, y_title3, y_title4]

    # figure with 4 subplots
    fig, axs = plt.subplots(2, 2, figsize=(8, 6), sharex=False, sharey=False)

    for x, data in enumerate(datas):
        i = x // 2
        h = x % 2
        # Get sample names and data
        x_labels = all_x_labels[x]
        x_labs = np.arange(len(x_labels))
        bottom = np.zeros(len(x_labels))
        # Create stacked bars
        for j, category in enumerate(legend_labels):
            bars = axs[i, h].bar(x_labs, data[:, j], bottom=bottom, edgecolor=plot_colour, width=0.9, 
                    label=category, color=colour_map[category])
            bottom += data[:, j].astype(float)
            if category == 'Classified Reads':
                for bar, value in zip(bars, data[:, j]):
                    axs[i, h].text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
                                   f'{value:.0f}%', ha='center', va='bottom', color=colour_map['Classified Reads'], 
                                   fontsize=12, fontweight='bold')

        # Plot titles, axis titles, axis labels
        axs[i, h].set_xticks(x_labs, x_labels, rotation=0)
        axs[i, h].set_xlabel(x_titles[x], color=font_colour)
        axs[i, h].set_ylabel(y_titles[x], color=font_colour, fontweight='bold')
        axs[i, h].set_title(plot_titles[x], color=font_colour)

    # # Aesthetics
    # Overall plot
    sub_fig_label = ['A)', 'B)', 'C)', 'D)']
    fig.suptitle(title, fontweight='bold', fontsize=14, color=font_colour)
    plt.tight_layout()
    fig.set_facecolor(background_colour)
    spine_color = line_colour
    spine_linewidth = 1.2
    # iterate over sub plots
    for i, ax_row in enumerate(axs):
        for ax in ax_row:
            ax.set_facecolor(plot_colour)
            for spine in ax.spines.values():
                spine.set_color(spine_color)
                spine.set_linewidth(spine_linewidth)
            ax.tick_params(axis='x', colors=line_colour)
            ax.tick_params(axis='y', colors=line_colour)
            ax.text(-0.1, 1.1, sub_fig_label[i], transform=ax.transAxes, fontsize=12, fontweight='bold', va='top', color=font_colour)
    
    # legend
    legend = axs[1,0].legend(loc='upper left', fontsize=10, 
                             ncols=2, bbox_to_anchor=(0.5, -0.275))
    for text in legend.get_texts():
        text.set_color(font_colour)

    # # Adjust layout to make room for the legend
    # plt.subplots_adjust(right=0.8)

    # # show plot
    plt.show()
```

```python
# FIGURE 1: PLOT
plot_title1 = 'Short Reads - No Preprocessing'
x_labels1 = ['A', 'B', 'C', 'E', 'F']
x_title1 = 'Mock'
y_title1 = '% of Reads'
plot_title2 = 'Quality Filtered Short Reads'
x_labels2 = ['A', 'B', 'C', 'E', 'F']
x_title2 = x_title1
y_title2 = y_title1
plot_title3 = 'Long Reads - No Preprocessing'
x_labels3 = ['A', 'B', 'C', 'D', 'E', 'F']
x_title3 = x_title1
y_title3 = y_title1
plot_title4 = 'Quality Filtered Long Reads'
x_labels4 = ['A', 'B', 'C', 'D', 'E', 'F']
x_title4 = x_title1
y_title4 = y_title1
title = 'Proportion of Classified Bacterial Reads'
legend_labels = ['Classified Reads', 'Unclassified Reads']
colour_map = {'Classified Reads': '#4CAF50', 'Unclassified Reads': '#E91E63'}

# colours 1
background_colour = '#D7F1FF'
plot_colour = '#FEF995'
font_colour = '#1A237E'
line_colour = '#1A237E'
figure_1(nopreSR_classunclass_df, filtSR_classunclass_df, 
             nopreLR_classunclass_df, filtLR_classunclass_df, 
             background_colour, plot_colour, font_colour, line_colour,
             title, legend_labels, colour_map,
             plot_title1, x_labels1, x_title1, y_title1,
             plot_title2, x_labels2, x_title2, y_title2,
             plot_title3, x_labels3, x_title3, y_title3,
             plot_title4, x_labels4, x_title4, y_title4)

# colours 2
background_colour = 'none'
plot_colour = 'whitesmoke'
font_colour = 'black'
line_colour = 'black'
figure_1(nopreSR_classunclass_df, filtSR_classunclass_df, 
             nopreLR_classunclass_df, filtLR_classunclass_df, 
             background_colour, plot_colour, font_colour, line_colour,
             title, legend_labels, colour_map,
             plot_title1, x_labels1, x_title1, y_title1,
             plot_title2, x_labels2, x_title2, y_title2,
             plot_title3, x_labels3, x_title3, y_title3,
             plot_title4, x_labels4, x_title4, y_title4)

# colours 3
background_colour = 'none'
plot_colour = 'none'
font_colour = 'black'
line_colour = 'black'
figure_1(nopreSR_classunclass_df, filtSR_classunclass_df, 
             nopreLR_classunclass_df, filtLR_classunclass_df, 
             background_colour, plot_colour, font_colour, line_colour,
             title, legend_labels, colour_map,
             plot_title1, x_labels1, x_title1, y_title1,
             plot_title2, x_labels2, x_title2, y_title2,
             plot_title3, x_labels3, x_title3, y_title3,
             plot_title4, x_labels4, x_title4, y_title4)
```
